# Stack and Unstack functions

Athletes wininng medals in Beijing Olympcs 100 or 200m track evet

```python
mw = oo[(oo.Edition == 2008) & ( (oo.Event == '100m') | (oo.Event =='200m'))]
```

```python
g = mw.groupby(['NOC','Gender','Discipline','Event']).size()
g
```

```python
df = g.unstack(['Discipline','Event'])
df
```

```python
df.stack(['Discipline'])
```

# Challenge

```python
import matplotlib.pyplot as plt
%matplotlib inline
gm = oo[oo.Medal == "Gold"]
gm.groupby(['Edition','Gender']).size()
```

```python
gm.groupby(['Edition','Gender']).size().unstack('Gender', fill_value=0)
gm.groupby(['Edition','Gender']).size().unstack('Gender', fill_value=0).plot(title='Gold Medals', kind='bar')
```

```python
# the 5 Athlete who have the most gold medals over the olympic games
gm = oo[oo.Medal == "Gold"]
gm.groupby(['Athlete','Medal']).size()
```

```python
g = oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head().plot(kind='bar')
```

## Creating own colormaps

```python
g = oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g
```

```python
from matplotlib.colors import ListedColormap
```

```python
sns.color_palette()
```

```python
gsb = ['#dbb40c','#c5c9c7','#a87900']
sns.palplot(gsb)
my_gsb = ListedColormap(sns.color_palette(gsb))
```
