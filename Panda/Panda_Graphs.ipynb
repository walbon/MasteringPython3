# How to use the graph plotting
## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import pandas as pd
# read the olimpics.csv
oo = pd.read_csv('./data/olympics.csv', skiprows=4)
```

```python
fo = oo[oo.Edition == 1896 ]
fo.head()
```

```python
fo.Sport.value_counts().plot(kind='line');
```

```python
fo.Sport.value_counts().plot(kind='bar');
```

```python
fo.Sport.value_counts().plot(kind='barh');
```

```python
fo.Sport.value_counts().plot(kind='pie');
```

```python
fo.Sport.value_counts().plot(kind='pie', colormap='Pastel2');
```

## Seaborn

```python
import seaborn as sns
```

```python
sns.countplot(x="Medal", data=oo);
```

```python
sns.countplot(x="Medal", data=oo, hue='Gender');
```

## Comparison

```python
chn = oo[ (oo.Edition == 2008) & (oo.NOC == "CHN")]
chn['Gender'].value_counts().plot(kind='pie')
```

```python
sns.countplot(x="Gender", data=oo[(oo.Edition==2008) & (oo.NOC == "CHN")]);
```

```python
sns.countplot(x="Medal", data=oo[(oo.Edition==2008) & (oo.NOC == "CHN")], hue='Gender', order=["Gold","Silver","Bronze"]);
```
